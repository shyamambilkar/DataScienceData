```python
Feature = df[['ENGINESIZE',
             'CYLINDERS',
             'FUELCONSUMPTION_CITY',
             'FUELCONSUMPTION_HWY',
             'FUELCONSUMPTION_COMB',
             'FUELCONSUMPTION_COMB_MPG']]
Feature
```

```python
target = df['CO2EMISSIONS']
```

```python
target[0:5]
```

# Train The Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(Feature, target, test_size = 0.2)
```

# Multiple Linear Regression

y = b0 + b1 * x1 + b2 * x2 + .....+bn *xn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
```

```python
model
```

### Fitting the Data

```python
model.fit(x_train, y_train)
```

### Slope of straight line

```python
model.coef_
```

### Intercept point in Y axis

```python
model.intercept_
```

```python
model.predict([[2.0,4,9.9,6.7,8.5,33]])
```

```python
df.head(1)
```

```python
new_pred = model.predict(x_test)
```

### Error = Actual - Predicted

```python
df1 = pd.DataFrame({"Actual_Data":y_test, "New Prediction":new_pred, "Error":y_test-new_pred})
```

```python
df1
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

**Total Mean Error**

```python
mean_squared_error(y_test, new_pred)
```

```python
mean_absolute_error(y_test, new_pred)
```

```python
train_score = model.score(x_train, y_train)
```

```python
print("Training score of Model is : {}".format(train_score))
```

```python
test_score = model.score(x_test, y_test)
print("Test score of model is :{}".format(test_score))
```
