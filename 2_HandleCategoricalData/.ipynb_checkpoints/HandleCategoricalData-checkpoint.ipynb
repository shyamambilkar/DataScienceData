```python
import pandas as pd
import datetime
```

1. Create variable with dates and from the extract the weekday
2. I create list of dates with 20 days difference from today
3. and then transform it into a datetime

```python
df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

**Extract the week day name**

```python
df['day_of_week'] = df['day'].dt.name
df
```

**Engineer categorical Variable by ordinal number replacement**

```python
weekday_map = {'Monday':1,
              'Tuesday':2,
              'Wednsday':3,
              'Thursday':4,
              'Friday':5,
              'Saturday':6,
              'Sunday':7}

df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head(20)
    
```
